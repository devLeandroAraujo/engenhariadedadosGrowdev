```python
import pandas as pd
```

```python
educ = 'C:\engenhariadedadosGrowdev\engenhariadedadosGrowdev\Tarefa8educ.csv'
gdp = 'https://raw.githubusercontent.com/abnr/ml-data/main/India_World_Bank_Info.csv'
```

```python
df_educ = pd.read_csv(educ, header=1, sep='\t', encoding= 'unicode_escape')
df_educ
```

```python
df_educ.info()
```

```python
df_educ.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
df_educ = df_educ[df_educ['Country'] == 'India']
series = ['Students enrolled in primary education (thousands)',
          'Students enrolled in secondary education (thousands)',
          'Students enrolled in tertiary education (thousands)']
gdp = df_educ['Series'].isin(series)
df_educ = df_educ[gdp]
df_educ
```

```python
df_educ['Value'] = df_educ['Value'].astype(str).str.replace(',','').astype(int)
df_educ
```

```python
df_educ.drop(['Region/Country/Area', 'Country', 'Source', 'Footnotes', 'Series'], axis=1, inplace=True)
df_educ
```

```python
df_educ = df_educ.groupby(['Year']).sum().astype(int)
df_educ
```

```python
df_gdp = pd.read_csv(gdp, sep='\t', on_bad_lines='error', skiprows=4)
df_gdp
```

```python
df_gdp.info()
```

```python
filtro_gdp = df_gdp['Indicator Name'].str.contains('GDP per capita \\(current US\\$\\)')
df_gdp = df_gdp[filtro_gdp]
df_gdp
```

```python
df_melted = pd.melt(df_gdp, id_vars=['Country Name', 'Country Code', 'Indicator Name'], 
                    var_name='Year', value_name='Value')

df_melted['Year'] = df_melted['Year'].astype(int) # Convertendo a coluna 'Year' para inteiro

# print(df_melted)
df_melted
```

```python
df_gdp_filtrado = df_melted[(df_melted['Year'] >= 2003) & (df_melted['Year'] <= 2018)]
df_gdp_final = df_gdp_filtrado[['Year', 'Value']]

# print(df_gdp_final)
df_gdp_final
```

```python
df_gdp_final.loc[:, 'Value'] = df_gdp_final['Value'].round().astype(int)

```

```python
df_gdp_final
```

```python
df_educ
```

```python
df_merged = pd.merge(df_gdp_final, df_educ, on='Year', suffixes=('_gdp', '_educ'), how='outer')
print(df_merged)
```

```python
print(df_merged['Value_educ'].dtype)
```

```python
df_merged['Value_educ'] = df_merged['Value_educ'].fillna(0).round().astype(int)
df_merged
```

```python
df_merged['Value_educ'] = pd.to_numeric(df_merged['Value_educ'], errors='coerce')
```

```python
print(df_merged['Value_educ'].dtype)
```

```python
df_merged['Value_educ'] = df_merged['Value_educ'].fillna(0).round().astype(int)
df_merged
```

```python
print(df_merged['Value_educ'].dtype)
print(df_merged['Value_gdp'].dtype)
```

```python
media_value_educ = df_merged['Value_educ'].replace(0, pd.NA).mean()

# Substituindo os zeros na coluna 'Value_educ' pela mÃ©dia calculada
df_merged['Value_educ'] = df_merged['Value_educ'].replace(0, media_value_educ)

# Dividindo a coluna 'Value_educ' por 10 e multiplicando a coluna 'Value_gdp' por 10
df_merged['Value_educ'] /= 10
df_merged['Value_gdp'] *= 10

# Exibindo o DataFrame atualizado
print("\nDataFrame atualizado:")
print(df_merged)
```

```python
print(df_merged['Value_educ'].dtype)
print(df_merged['Value_gdp'].dtype)
```
