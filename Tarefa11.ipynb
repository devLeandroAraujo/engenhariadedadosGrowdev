1. Selecione todos produtos de uma categoria

```python

categoria_especifica = 'Eletrônico'  # Substitua pelo nome da categoria desejada ()
query1 = f"""
    SELECT * FROM Produtos
    WHERE Categoria = '{categoria_especifica}';
"""
df_categoria = pd.read_sql(query1, engine)
print(f"1. Produtos da {categoria_especifica}:")
df_categoria
```

2. Encontre o nome e o endereço dos clientes que fizeram um pedido em 2024

```python
query2 = """
    SELECT Clientes.Nome, Clientes.Endereco
    FROM Clientes
    JOIN Pedidos ON Clientes.ID_Cliente = Pedidos.ID_Cliente
    WHERE EXTRACT(YEAR FROM Pedidos.Data_Pedido) = 2024;
"""
df_clientes_2024 = pd.read_sql(query2, engine)
print("\n2. Nome e endereço dos clientes com pedidos em 2024:")
df_clientes_2024
```

3. Liste os produtos cujo preço está entre 50 e 100

```python
query3 = """
    SELECT * FROM Produtos
    WHERE Preco BETWEEN 50 AND 100;
"""
df_preco_intervalo = pd.read_sql(query3, engine)
print("\n3. Produtos com preço entre 50 e 100:")
df_preco_intervalo
```

4. Conte o número total de pedidos feitos no mês de julho de 2024

```python
query4 = """
    SELECT COUNT(*) AS total_pedidos_julho
    FROM Pedidos
    WHERE EXTRACT(YEAR FROM Data_Pedido) = 2024
    AND EXTRACT(MONTH FROM Data_Pedido) = 7;
"""
df_pedidos_julho = pd.read_sql(query4, engine)
print("\n4. Total de pedidos feitos em julho de 2024:")
df_pedidos_julho
```

5. Encontre o total pago por cada pedido

```python
query5 = """
    SELECT ID_Pedido, SUM(Valor_Pago) AS total_pago
    FROM Pagamentos
    GROUP BY ID_Pedido;
"""
df_total_pago_pedido = pd.read_sql(query5, engine)
print("\n5. Total pago por cada pedido:")
df_total_pago_pedido
```

6. Liste o nome dos clientes e o valor total dos seus pedidos

```python
query6 = """
    SELECT Clientes.Nome, SUM(Pedidos.Valor_Total) AS valor_total_pedidos
    FROM Clientes
    JOIN Pedidos ON Clientes.ID_Cliente = Pedidos.ID_Cliente
    GROUP BY Clientes.Nome;
"""
df_total_pedidos_cliente = pd.read_sql(query6, engine)
print("\n6. Nome dos clientes e valor total dos pedidos:")
df_total_pedidos_cliente
```

7. Encontre os produtos com estoque menor que 10

```python
query7 = """
    SELECT * FROM Produtos
    WHERE Estoque < 10;
"""
df_estoque_baixo = pd.read_sql(query7, engine)
print("\n7. Produtos com estoque menor que 10:")
df_estoque_baixo
```

8. Liste os pedidos com seu status e a quantidade de itens em cada pedido

```python
query8 = """
    SELECT Pedidos.ID_Pedido, Pedidos.Status, COUNT(Itens_Pedido.ID_Item) AS quantidade_itens
    FROM Pedidos
    JOIN Itens_Pedido ON Pedidos.ID_Pedido = Itens_Pedido.ID_Pedido
    GROUP BY Pedidos.ID_Pedido, Pedidos.Status;
"""
df_itens_pedido = pd.read_sql(query8, engine)
print("\n8. Pedidos com status e quantidade de itens:")
df_itens_pedido
```

9. Encontre os produtos que não foram vendidos em nenhum pedido

```python
query9 = """
    SELECT Produtos.Nome
    FROM Produtos
    LEFT JOIN Itens_Pedido ON Produtos.ID_Produto = Itens_Pedido.ID_Produto
    WHERE Itens_Pedido.ID_Produto IS NULL;
"""
df_produtos_nao_vendidos = pd.read_sql(query9, engine)
print("\n9. Produtos que não foram vendidos em nenhum pedido:")
df_produtos_nao_vendidos
```

10. Liste o nome dos 10 maiores clientes e a quantidade de pedidos feitos por cada um

```python
query10 = """
    SELECT Clientes.Nome, COUNT(Pedidos.ID_Pedido) AS quantidade_pedidos
    FROM Clientes
    JOIN Pedidos ON Clientes.ID_Cliente = Pedidos.ID_Cliente
    GROUP BY Clientes.Nome
    ORDER BY quantidade_pedidos DESC
    LIMIT 10;
"""
df_top_clientes = pd.read_sql(query10, engine)
print("\n10. Top 10 clientes por quantidade de pedidos:")
df_top_clientes
```

11. Encontre o produto mais caro e o mais barato

```python
query11 = """
   (
    SELECT Nome, Preco
    FROM Produtos
    ORDER BY Preco DESC
    LIMIT 1
)
UNION ALL
(
    SELECT Nome, Preco
    FROM Produtos
    ORDER BY Preco ASC
    LIMIT 1
);
"""
df_preco_extremos = pd.read_sql(query11, engine)
print("\n11. Produto mais caro e mais barato:")
df_preco_extremos
```

12. Calcule a idade média dos clientes

```python
query12 = """
    SELECT AVG(EXTRACT(YEAR FROM AGE(Data_Nascimento))) AS idade_media
    FROM Clientes;
"""
df_idade_media = pd.read_sql(query12, engine)
print("\n12. Idade média dos clientes:")
df_idade_media
```

13. Liste os clientes que possuem o mesmo sobrenome

```python
query13 = """
    SELECT Nome, COUNT(*)
    FROM Clientes
    GROUP BY Nome
    HAVING COUNT(*) > 1;
"""
df_sobrenome_repetido = pd.read_sql(query13, engine)
print("\n13. Clientes com o mesmo sobrenome:")
df_sobrenome_repetido
```

14. Encontre os pedidos feitos por clientes cujo nome começa com a letra 'A'

```python
# Abre uma conexão bruta para o SQL
connection = engine.raw_connection()
try:
    query14 = """
        SELECT Pedidos.*
        FROM Pedidos
        JOIN Clientes ON Pedidos.ID_Cliente = Clientes.ID_Cliente
        WHERE Clientes.Nome LIKE 'A%'
    """
    
    # Executa a consulta e lê os dados em um DataFrame
    df_pedidos_nome_a = pd.read_sql_query(query14, con=connection)
    print("14. Pedidos de clientes cujo nome começa com 'A':")
    print(df_pedidos_nome_a)
finally:
    connection.close()  # Certifique-se de fechar a conexão após o uso

```

15. Liste os produtos, sua categoria e uma indicação se estão em estoque ou esgotados (usando CASE)

```python
query15 = """
    SELECT Nome, Categoria,
    CASE
        WHEN Estoque > 0 THEN 'Em Estoque'
        ELSE 'Esgotado'
    END AS status_estoque
    FROM Produtos;
"""
df_estoque_status = pd.read_sql(query15, engine)
print("\n15. Produtos com status de estoque:")
df_estoque_status
```
