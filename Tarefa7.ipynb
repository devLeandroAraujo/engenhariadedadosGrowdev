#Exercícios pandas e plot

As questões a seguir utilizam como base o dataset de venda de jogos eletrônicos disponível no seguinte [link.](https://www.kaggle.com/code/rafa84miranda/vendas-de-jogos-de-videogames/data?select=vgsales.csv)

Os campos são:
- Rank: posição no ranking de vendas
- Name: nome do jogo
- Platform: plataforma/console
- Year: ano de lançamento
- Genre: gênero
- Publisher: empresa que publicou o jogo
- NA_Sales: vendas na américa do norte (em milhões de dólares)
- EU_Sales: vendas na europa (em milhões de dólares)
- JP_Sales: vendas no japão (em milhões de dólares)
- Other_Sales: outras vendas (em milhões de dólares)
- Global_Sales: vendas globais (em milhões de dólares)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arquivo_csv = 'C:\engenhariadedadosGrowdev\engenhariadedadosGrowdev\Tarefa7vgsales.csv'
df_vgsales = pd.read_csv(arquivo_csv)
df_vgsales
```

```python
df_vgsales.describe().T
```

Qual foi o jogo mais vendido no mundo em 2010?



```python
jogos_2010 = df_vgsales[df_vgsales['Year'] == 2010]

jogo_mais_vendido_2010 = jogos_2010.loc[jogos_2010['Global_Sales'].idxmax()]

print(jogo_mais_vendido_2010)
```

Qual foi a empresa que mais vendeu de 2005 até 2016 - Construa um gráfico de barras para apresentar


```python
jogos_2005_2016 = df_vgsales[(df_vgsales['Year'] >= 2005) & (df_vgsales['Year'] <= 2016)]

vendas_por_empresa = jogos_2005_2016.groupby('Publisher')['Global_Sales'].sum().reset_index()

vendas_por_empresa = vendas_por_empresa.sort_values(by='Global_Sales', ascending=False)

top_empresas = vendas_por_empresa.head(10)

# Criando o gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(top_empresas['Publisher'], top_empresas['Global_Sales'], color='Red')
plt.title('Top 10 Empresas com Mais Vendas (2005-2016)')
plt.xlabel('Publisher')
plt.ylabel('Vendas Globais (milhões de dólares)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Qual o nome do jogo que menos foi vendido em 2008 do gênero Sports.


```python
jogos_2008_sports = df_vgsales[(df_vgsales['Year'] == 2008) & (df_vgsales['Genre'] == 'Sports')]

jogo_menos_vendido_2008 = jogos_2008_sports.loc[jogos_2008_sports['Global_Sales'].idxmin()]

print(jogo_menos_vendido_2008[['Name', 'Global_Sales']])
```

Qual o percentual relacionado às vendas globais de jogos de Wii (excluindo-se América do norte, japão e europa) de 2006 até 2010?





```python
# jogos Wii entre 2006 e 2010
wii_games = df_vgsales[(df_vgsales['Platform'] == 'Wii') & (df_vgsales['Year'].between(2006, 2010))]

# outras regiões
total_other_sales = wii_games['Other_Sales'].sum()

#  vendas globais
total_global_sales = wii_games['Global_Sales'].sum()

percentual_other_sales = (total_other_sales / total_global_sales) * 100 if total_global_sales > 0 else 0

print(f'O percentual das vendas globais de jogos de Wii (excluindo América do Norte, Japão e Europa) de 2006 até 2010 é: {percentual_other_sales:.2f}%')

```

Mostre um gráfico de linhas com as vendas em diferentes plataformas ao longo dos anos.


```python
# Agrupando os dados por Year e Platform e somando as vendas globais
vendas_por_plataforma = df_vgsales.groupby(['Year', 'Platform'])['Global_Sales'].sum().reset_index()

# Criando uma tabela dinâmica para reorganizar os dados
pivot_vendas = vendas_por_plataforma.pivot(index='Year', columns='Platform', values='Global_Sales').fillna(0)

# Criando o gráfico de linhas
plt.figure(figsize=(14, 7))
for platform in pivot_vendas.columns:
    plt.plot(pivot_vendas.index, pivot_vendas[platform], label=platform)

plt.title('Vendas Globais de Jogos por Plataforma ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Vendas Globais (milhões de dólares)')
plt.legend(title='Plataforma', bbox_to_anchor=(0.5, -0.1), loc='upper center', ncol=10)
plt.grid()
plt.tight_layout()
plt.show()
```

Mostre um gráfico de pizza com o percentual de vendas dos diferentes gêneros de jogos.




```python
vendas_por_genero = df_vgsales.groupby('Genre')['Global_Sales'].sum()

plt.figure(figsize=(10, 10))
plt.pie(vendas_por_genero, labels=vendas_por_genero.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentual de Vendas por Gênero de Jogos')
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

Faça a concatenação das bases de dados livros.csv e livros_genero.csv e encontre qual o gênero de livro correspondente a cada livro.

```python
mapa_genero = {
    'Pedro': 'Drama',
    'Maria': 'Romance',
    'Carlos': 'Drama',
    'Fabio': 'Romance',
    'Aline': 'Aventura',
    'Sandro': 'Aventura'
}
df_livros_classicos['genero'] = df_livros_classicos['autor'].map(mapa_genero)

df_livros_completos = pd.merge(df_livros, df_livros_classicos, on='nome')


df_livros_completos
```

Faça a concatenação das bases de dados de livros e livros_genero e encontre qual os gêneros e os autores que não possuem um livro correspondente.

```python
df_livros_genero['livro'] = None 

df_concatenado = pd.concat([df_livros_genero, df_livros.rename(columns={'nome': 'livro'})], ignore_index=True)

generos_com_livros = set(df_livros_genero['genero'])
livros_com_genero = set(df_concatenado['genero'].dropna())

generos_sem_livros = generos_com_livros - livros_com_genero

df_generos_sem_livros = pd.DataFrame({'genero': list(generos_sem_livros)})

autores_com_livros = set(df_livros_genero['autor'])
livros_nomes = set(df_livros['nome']) 

autores_sem_livros = autores_com_livros - livros_nomes

df_autores_sem_livros = pd.DataFrame({'autor': list(autores_sem_livros)})

print("Gêneros que não possuem um livro correspondente:")
df_generos_sem_livros

print("\nAutores que não possuem um livro correspondente:")
df_autores_sem_livros
```

Faça a junção das bases de livros e livros clássicos apenas com os elementos comuns entre ambos, e adicione chaves para o dataframe de livros e de livros clássicos, além de nomes para os índices.

Faça a junção das bases de dados livros e livros_classicos e verifique a integridade do dataframe resultante.

```python
df_juncao = pd.merge(df_livros, df_livros_classicos, on='nome', how='inner')

# Verificar a integridade
integridade = {
    'Linhas duplicadas': df_juncao.duplicated().sum(),
    'Valores nulos': df_juncao.isnull().sum().sum(),
    'Total de linhas': len(df_juncao),
    'Total de colunas': df_juncao.shape[1]
}

print("DataFrame após a junção:")
print(df_juncao)

print("Verificação de integridade:")
for key, value in integridade.items():
    print(f"{key}: {value}")
```
